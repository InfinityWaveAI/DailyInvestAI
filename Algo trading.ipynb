```python
## NSE data :>
import nsepython

symbol = "NIFTY 50"
symbol = "SBIN"
start_date = "1-May-2024"
end_date = "1-Nov-2024"
nifty = nsepython.index_history(symbol,start_date,end_date)
```

```python
nifty.shape
```

```python
adj_price = nifty['CLOSE'].astype(float)
adj_price.shape
```

```python
no_days = 7
# calculate the moving average
mav = adj_price.rolling(window=no_days).mean()
 
```

```python
# import the matplotlib package 
# to see the plot
import matplotlib.pyplot as plt

adj_price.plot()
mav.plot()
```

```python
## Equity
symbol = "ADANIENT"
series = "EQ"
start_date = "04-06-2024"
end_date ="27-12-2024"
temp = nsepython.equity_history(symbol,series,start_date,end_date)
temp.shape
```

```python
temp
```

```python
adj_price = temp['CH_CLOSING_PRICE'].astype(float)
adj_price.shape
```

```python
no_days = 7
# calculate the moving average
mav = adj_price.rolling(window=no_days).mean()
 
```

```python
# import the matplotlib package 
# to see the plot
import matplotlib.pyplot as plt

adj_price.plot()
mav.plot()
```
